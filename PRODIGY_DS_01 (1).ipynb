```
df = '/content/Populationdata.csv'
data = pd.read_csv(df, skiprows = 3 , header = 1)
data.drop(['Indicator Name', 'Indicator Code','Country Code'], axis = 1, inplace=True)
data.head()
```

```
Aruba_data = data[data['Country Name'] == 'Aruba']
```

```
Aruba_data
```

```
Aruba_data_file = Aruba_data.melt(id_vars='Country Name', var_name='Year', value_name='Population')
Aruba_data_file.drop(['Country Name'], axis = 1, inplace=True)
Aruba_data_file['Year'] = Aruba_data_file['Year'].astype(str)
Aruba_data_file['Population'] = Aruba_data_file['Population'].astype(str)

# print(Aruba_data_file['Population'].unique())
```

```
# Create the bar chart
plt.figure(figsize=(14, 8))  # Adjust figure size as needed
plt.bar(Aruba_data_file['Year'], Aruba_data_file['Population'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Aruba Population Over Time')  # Adjust title for Aruba data
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a faint grid for better visualization
plt.tight_layout()  # Adjust spacing between elements

# Display the chart
plt.show()
```
